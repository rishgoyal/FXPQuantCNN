```python
def plot(weights):
    fig, ax = plt.subplots(12, 2, figsize=(18, 40))
    ax = ax.flatten()
    for i in range(len(weights)):
        ax[i].hist(weights[i], bins=30, label=['weights', 'biases']);
        ax[i].set_title(layer_names[i])
        ax[i].legend()
    fig.delaxes(ax[-1])
```
